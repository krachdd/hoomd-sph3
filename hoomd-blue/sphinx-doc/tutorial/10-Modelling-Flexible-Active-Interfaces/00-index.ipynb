# Modelling Flexible Active Interfaces

This tutorial explains how to model deformable interfaces such as vesicles or membranes by using meshes. 

**Prerequisites:**

This tutorial assumes you are familiar with the concepts introduced in 
[Introducing HOOMD-blue](../00-Introducing-HOOMD-blue/00-index.ipynb),
[Introducing Molecular Dynamics](../01-Introducing-Molecular-Dynamics/00-index.ipynb) and
[Custom Actions in Python](../04-Custom-Actions-In-Python/00-index.ipynb).

## Outline

1. [Describing Membranes as Meshes](01-Describing-Membranes-As-Meshes.ipynb)
2. [Simulating a Vesicle with MD](02-Vesicle-MD.ipynb)
